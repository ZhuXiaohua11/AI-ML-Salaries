Korrelationsmatris， not obvious

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 10))
sns.heatmap(salaries_en_df.corr(), annot=False, cmap='coolwarm')
plt.title("Correlation Heatmap of Salary Dataset")
plt.show()
```

EDA – Utforskande analys
Histogram över löner

```python
plt.figure(figsize=(10, 6))
plt.hist(salaries_en_df['salary_in_usd'], bins=50, color='blue')
plt.title('Distribution of Salary in USD')
plt.xlabel('Salary in USD')
plt.ylabel('Antal personer')
plt.grid(True)
plt.show()
```

StandardScaler

```python
from sklearn.preprocessing import StandardScaler
# Skapa en StandardScaler
scaler = StandardScaler()

# Standardisera alla kolumner
# Obs: Här antar vi att hela salaries_en_df bara innehåller numeriska data!
salaries_scaled = scaler.fit_transform(salaries_en_df)

# Konvertera till DataFrame igen (om du vill)
import numpy as np
salaries_scaled_df = pd.DataFrame(
    salaries_scaled,
    columns=salaries_en_df.columns
)
```
